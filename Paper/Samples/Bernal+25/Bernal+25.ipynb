```python
with open("bernal_spectras.pkl", "rb") as f:
    dict_spectra = pickle.load(f)
base_panda = pd.read_csv("bernal_spectras.csv")
```

```python
names = list(dict_spectra.keys())
spectra = np.stack([i["new"] for i in dict_spectra.values()])
z = np.stack([i["z"] for i in dict_spectra.values()])
coords = np.stack([i["coords"] for i in dict_spectra.values()])
```

```python
sheapspectral = Sheapectral(spectra,z=z,coords=coords,names=names)
```

```python
add_host_miles = {"z_include":None,"age_include":None}
sheapspectral.makemodel((3800,8000),n_broad=2,n_narrow=1,group_method= True ,fe_mode ="template",continuum_profile="powerlaw",
                          add_winds=False,
                            add_outflow=False,
                           add_uncommon_narrow=False,add_host_miles = add_host_miles,add_balmer_continuum=True,verbose=True)
```

```python
sheapspectral.fitmodel(list_num_steps=[2_000],list_learning_rate=[1e-2],run_fit=True)
```
